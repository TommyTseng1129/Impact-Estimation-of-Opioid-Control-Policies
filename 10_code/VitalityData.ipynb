```python
# Dropping all observations that do not have a cause of death related to drug overdose

drugCodes = ['D1', 'D2', 'D4', 'D9']
drug_vitality1 = vitality[vitality['Drug/Alcohol Induced Cause Code'].isin(drugCodes)]

drug_vitality1['Drug/Alcohol Induced Cause Code'].value_counts()
```

```python
# Rename columns to make them less cumbersome
drug_vitality2 = drug_vitality1.copy()
drug_vitality2.rename(columns = {'Drug/Alcohol Induced Cause Code':'Death Code', 'Drug/Alcohol Induced Cause':'Cause of Death'}, inplace = True) 
```

```python
drug_vitality3 = drug_vitality2.copy()
drug_vitality3
```

```python
#Drog extraneous column "Year Code" which is the same as year, also drop the last line of data in the framew which is a wierd na/year value that was accidently added

drug_vitality3.drop(['Year Code'], axis=1, inplace = True)
#drug_vitality3.drop(drug_vitality.tail(1).index,inplace=True) 
```

```python
# Check the Data frame for Null values 

drug_vitality3['County'].isnull().values.any()
drug_vitality3['County Code'].isnull().values.any()
drug_vitality3['Year'].isnull().values.any()
drug_vitality3['Cause of Death'].isnull().values.any()
drug_vitality3['Death Code'].isnull().values.any()
drug_vitality3['Deaths'].isnull().values.any()
```

```python
drug_vitality3[drug_vitality3['County Code'] == 24003]
```

```python
# Create a copy of the final data set to create version with total deaths per county per year, not taking type of overdose into account
import numpy as np

od_deaths = drug_vitality3.copy()
od_deaths.replace('Missing', np.nan, inplace=True)
od_deaths['Deaths'] = od_deaths['Deaths'].astype(float)
#print(od_deaths.dtypes)
```

```python
od_deaths
```

```python
#Drop drug cause of death and code

od_deaths.drop(['Cause of Death', 'Death Code'], axis = 1, inplace = True)
od_deaths[od_deaths['County Code'] == 24003]
```

```python
#Groupby County and Year 

od_deaths = od_deaths.groupby(["County","County Code", "Year"], as_index = False).sum()
```

```python
# Final Dataset

od_deaths['Year'].value_counts()
od_deaths[od_deaths['County Code'] == 24003]

#od_deaths['State'] = 
od_deaths_state = od_deaths['County'].str.split(pat=', ', n=1, expand=True)
```

```python
od_deaths_state['County Code'] = od_deaths['County Code']
od_deaths_state['Year'] = od_deaths['Year']
od_deaths_state['Deaths'] = od_deaths['Deaths']

od_deaths_state
```

```python
od_deaths_state.rename(columns = {0:'County', 1:'State'}, inplace = True) 
od_deaths_state
```
